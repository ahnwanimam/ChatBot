```python
import urllib.request
import pandas as pd

#URL에서 데이터 다운로드 하여 로컬파일 저장 / URL : 일반 채팅 데이터
urllib.request.urlretrieve("https://raw.githubusercontent.com/songys/Chatbot_data/master/ChatbotData.csv", filename="Gen_ChatData.csv")
```

```python
#저장한 csv 가져오기
Gen_ChatData = pd.read_csv('Gen_ChatData.csv')
print(Gen_ChatData.tail())

#일반 채팅 데이터 불필요한 null값 확인
print(Gen_ChatData.isnull().sum())
```

```python
# 원하는 데이터프레임으로 추출
Gen_ChatData = Gen_ChatData[['Q', 'A']]
#DF 확인
Gen_ChatData
```

```python
#서경 데이터 엑셀 파일을 데이터프레임으로 로드
SKChatData = pd.read_excel('C:\\Users\\a\\Desktop\\NLP\\QnA_pt\\SKChatData.xlsx')

#원하는 데이터프레임으로 추출
SKChatData = SKChatData[['Q', 'A']]
#DF 확인
SKChatData
```

```python
#일반채팅 데이터와 서경데이터 합치기
final_Data = pd.concat([Gen_ChatData, SKChatData])

final_Data
#CSV 파일로 저장
final_Data.to_csv('SK_finalData.csv', index=False)
```

```python
import torch
from sentence_transformers import SentenceTransformer
```

```python
#개발자 모드로 인한 에러를 방지하기 위한 조치
import os
os.environ["HF_HUB_DISABLE_SYMLINKS_WARNING"] = "1"

#선학습 된 KR-SBERT 모델 로드 
model = SentenceTransformer('snunlp/KR-SBERT-V40K-klueNLI-augSTS')
```

```python
#progress_map을 사용하기 위해 Pandas DataFrame과 Series 객체에 진행률 표시 기능을 추가
from tqdm import tqdm
tqdm.pandas()

#질문데이터 임베딩
final_Data['embedding_vector'] = final_Data['Q'].progress_map(lambda x : model.encode(x))
```

```python
#임베딩 데이터 확인
print(final_Data['embedding_vector'])
print(final_Data['embedding_vector'].dtype)
```

```python
#텐서화하기 위해 Numpy배열 사용
import numpy as np

final_embedding_data = final_Data['embedding_vector'].tolist()
print(final_embedding_data[0])
final_embedding_data_np = np.array(final_embedding_data, dtype=np.float32)
final_embedding_data_tensor = torch.tensor(final_embedding_data_np)

final_embedding_data_tensor
```

```python
#질문 데이터 임베딩 파일 저장
torch.save(final_embedding_data_tensor, 'SKChat_final_embedding_data_tensor.pt')
```
